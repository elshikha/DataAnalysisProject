```python
# Importing Libraries
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset without time column
df = df.drop("Timestamp",axis = 1)
df.head()
```

```python
# Checking for Null Values
df.isna().sum()
```

```python
# Average C-GPA
mean_cgpa = np.mean(df["What's your current C-GPA?"])
print(f'Mean: {mean_cgpa}')
# Median C-GPA
median_cgpa = np.median(df["What's your current C-GPA?"])
print(f'Median: {median_cgpa}')
# Mode C-GPA
mode_cgpa = stats.mode(df["What's your current C-GPA?"])
print(f'Mode: {mode_cgpa}')
```

```python
# Transforming categorical data into numerical for correlation
prep = np.zeros(34)
data = df["When it comes to preparing for exams or completing assignments, do you tend to study last minute or plan your study sessions well in advance?"]
for i in range(len(data)):
    if data[i] == "Often":
        prep[i] = 0 
    elif data[i] == "In advance":
        prep[i] = 1
prep
```

```python
# Correlation between C-GPA and Preparation
np.corrcoef(prep,df["What's your current C-GPA?"])[1][0]
```

```python
# Scatter Plot between C-GPA and Preparation
plt.scatter(df["What's your current C-GPA?"],prep,color='darkcyan')
plt.xlabel("C-GPA")
plt.ylabel("Preparation")
plt.title("Scatter Plot of C-GPA and Preparation Level")
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```

```python
# Transforming categorical data into numerical for correlation
procras = np.zeros(34)
data = df["How often do you procrastinate when it comes to studying?"]
for i in range(len(data)):
    if data[i] == "Never":
        procras[i] = 0
    elif data[i] == "Rarely":
        procras[i] = 1
    elif data[i] == "Sometimes":
        procras[i] = 2
    elif data[i] == "Often":
        procras[i] = 3
    elif data[i] == "Always":
        procras[i] = 4
procras
```

```python
# Correlation between C-GPA and Procastination
np.corrcoef(procras,df["What's your current C-GPA?"])[1][0]
```

```python
# Scatter Plot between C-GPA and Procastination
plt.scatter(df["What's your current C-GPA?"],procras,color='darkcyan')
plt.xlabel("C-GPA")
plt.ylabel("Procastination")
plt.title("Scatter Plot of C-GPA and Procastination Level")
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```
